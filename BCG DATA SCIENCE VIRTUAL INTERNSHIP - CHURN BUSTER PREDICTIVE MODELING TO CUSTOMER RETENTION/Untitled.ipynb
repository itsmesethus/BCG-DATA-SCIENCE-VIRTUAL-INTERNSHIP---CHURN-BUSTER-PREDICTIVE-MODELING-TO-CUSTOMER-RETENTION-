```python
ints=df.select_dtypes(include=['int','float'])
```

```python
q1=ints.quantile(0.25)
q3=ints.quantile(0.75)
iqr=q3-q1
up=q3+1.5*(iqr)
ll=q1-1.5*(iqr)
mask=((ints < ll)|(ints >up)).any(axis=1)
```

```python
df1=df[~mask]
df1
```

```python
mn=ints.mean()
s=ints.std()
z=(ints-mn)/s
thres=3
mask=np.abs(z<=3).all(axis=1)
df5=df[mask]
df5
```
